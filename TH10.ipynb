```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


img = cv2.imread('hehe.webp', 0)  # Đọc ảnh ở chế độ grayscale

# Tạo kernel (nhân) cho các phép toán hình thái học
kernel = np.ones((5,5), np.uint8)

# phép toán Erosion (Xói mòn)
erosion = cv2.erode(img, kernel, iterations = 1)

# phép toán Dilation (Giãn nở)
dilation = cv2.dilate(img, kernel, iterations = 1)


plt.subplot(131),plt.imshow(img, cmap = 'gray'),plt.title('Original')
plt.subplot(132),plt.imshow(erosion, cmap = 'gray'),plt.title('Erosion')
plt.subplot(133),plt.imshow(dilation, cmap = 'gray'),plt.title('Dilation')
plt.show()
```
